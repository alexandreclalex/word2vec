```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from CONFIG import PREPROCESSED_DATA_PATH
```

```python
df_data = pd.read_parquet(PREPROCESSED_DATA_PATH)
```

```python
df_data.info()
df_data["title"] = df_data["title"].astype(str)
df_data["categories"] = df_data["categories"].astype(str)
df_data.info()
```

```python
print("total product titles:", len(df_data))
```

```python
df_data.head()
```

### Category Analysis

```python
unique_cats = np.unique(df_data["categories"])
print("Number of Unique Categories:", len(unique_cats))
print("Average Products Per Category: ", int(len(df_data) / len(unique_cats)))
```

```python
plt.figure(figsize=(16, 6), dpi=80)
plt.title("Products Per Category")
plt.hist(df_data["categories"], rwidth=0.7, bins=len(unique_cats))
plt.xticks(rotation=90)
plt.xlabel("Category")
plt.ylabel("Num Products")
plt.show()
```

### Title Analysis

```python
lengths = np.array([len(x) for x in df_data["title"]])
print("Average Title Length:", np.sum(lengths) / len(df_data))
print("Shortest Title Length:", np.min(lengths))
print("Longest Title Length:", np.max(lengths))
```

```python
plt.title("Title Lengths")
plt.hist(lengths)
plt.xlabel("Character Length")
plt.ylabel("Num Titles")
```
